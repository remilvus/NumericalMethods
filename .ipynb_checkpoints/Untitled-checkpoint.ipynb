# SVD

```python
import numpy as np
from numpy.linalg import svd
from PIL import Image
import matplotlib.pyplot as plt
```

```python
lenna = Image.open("Lenna.png")
lenna_arr = np.array(lenna)
```

```python
agh = Image.open("agh.png").resize((512,512))
agh_arr = np.array(agh)[:,:,0:3]
```

```python
def compress_image(image, k):
    colors = [image[:,:,i] for i in range(3)]
    compressed = []
    for color in colors:
        u, s, vh = svd(color)
        Sigma = np.zeros((color.shape[0], color.shape[1]))
        Sigma[:color.shape[0], :color.shape[0]] = np.diag(s)
        compressed.append(u @ Sigma[:,:k] @ vh[:k,:])
    return np.array(np.stack(compressed, axis=2), dtype=np.uint8)
```

```python
def illustrate_compression(image):
    plt.figure(figsize=(15,25))
    for i, k in enumerate([500, 100, 50, 20, 10]):
        compressed = compress_image(image, k)
        plt.subplot(5, 2, 2*i+1)
        plt.axis("off")
        plt.title(f"compressed image, k={k}")
        plt.imshow(compressed)
        plt.subplot(5, 2, 2*i+2)
        plt.axis("off")
        diff = np.sum(abs(compressed.astype(np.int32)-image.astype(np.int32)), axis=2)
        plt.imshow(diff, cmap='gray')
        plt.title(f"mean difference = {np.mean(diff)}")
```

```python
illustrate_compression(lenna_arr)
```

```python
illustrate_compression(agh_arr)
```

```python
rand = np.random.randint(0, 255, (512,512,3))
illustrate_compression(rand)
```
